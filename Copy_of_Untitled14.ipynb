Data Binning and data formatting using python

```python
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'price':[55000,20000,30000,2000,8000,45000],
    'Unit_Sold':[25,60,35,80,50,15]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Data Binning
bins1 = [0,10000,30000,60000]
labels1=['Low','Medium','High']

df['Price_Category']=pd.cut(df['price'],bins=bins1,labels=labels1)
print(df)
```

```python
#Binning units sold
bins2=[0,30,60,100]
labels2 = ['Low Sales','Medium Sales','High Sales']
df['Sales_Category']=pd.cut(df['Unit_Sold'],bins=bins2,labels=labels2)
print(df)
```

```python
#Data Formatting
#Convert Data Type
print(df.dtypes)
df['Unit_Sold']=df['Unit_Sold'].astype(float)
print(df.dtypes)
print(df)
```

```python
#Convert text to uppercase
df['Product']=df['Product'].str.upper()
print(df)
```

```python
#Rpunding Values
df['Price']=df['price'].round(2)
print(df)
```

```python
#Additional Operations
#Sort Data
sorted_df=df.sort_values(by='Price')
print(sorted_df)
```

```python
df.sort_values(by='Unit_Sold',ascending=False,inplace=True)
print(df)
```

```python
#Display Unique Values
print(df['Price_Category'].unique())
print(df)
```

```python
#Dataset:Food Delivery Orders
import pandas as pd
import numpy as np

data = {
    'Order_ID':['01','02','03','04','05','06','07','08'],
    'Order_Value':[250,450,120,600,300,150,500,220],
    'Delivery_Time':[30,45,25,60,35,20,50,28],
    'Distance_km':[5,8,3,10,6,4,12,7]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Binning Order Value
bins = [0,200,400,700]
labels = ['Low','Medium','High']
df['Order_Category']=pd.cut(df['Order_Value'],bins=bins,labels=labels)
print(df)
```

```python
#Binning Delivery Time
bins2=[0,30,45,70]
labels2=['Fast','Medium','Slow']
df['Delivery_Category']=pd.cut(df['Delivery_Time'],bins=bins2,labels=labels)
print(df)
```

```python
#Binning Delivery_Km
bins3=[0,5,8,12]
labels3=['Near','Medium','Far']
df['Distance_Category']=pd.cut(df['Distance_km'],bins=bins3,labels=labels3)
print(df)
```
